# Coin Flip Simulator


A python code to flip a coin multiple times.
Also uses the Image module to display the outcome of the flip.

```python
#MODULES NEEDED
from PIL import Image
import random
import time

#A function is defined
def coinflip():
    play=True
    while play:
        #Variables are assigned for both images
        x=Image.open('head.png')
        y=Image.open('tails.png')
        
        #Random number for each flip
        flip=random.randint(0,1)
        #Optional wait time for the first flip
        time.sleep(0.5)
        if flip==0:
            display(x)
            print("It's HEADS !!!!!")
        elif flip==1:
            display(y)
            print("It's TAILS !!!!")
        
        #For Continous flips
        choice=input("Do you wish to flip again? Enter (y/n)").lower()
        #Wait time between every flip
        time.sleep(0.8)
        if choice=='y':
            play=True
        elif choice=='n':
            play=False

coinflip()
        
```
